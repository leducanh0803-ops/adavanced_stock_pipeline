```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
spark = SparkSession.builder.appName('test').getOrCreate()
df = spark.read.parquet("s3a://lakehouse/silver/ohlcv_74d1f9a7-7f19-47ca-8f7f-88ccfcddc94b/data")
```

```python
##Sort 
symbol_sort = df.orderBy(F.col('symbol').desc())
symbol_sort.show(2)
```

```python
df_e =  df.filter(F.col('symbol').isin(['AAPL','MSFT','TSLA','MMM']))
```

```python
df_e.show()
```

```python
from pyspark.sql import Window
window = Window.partitionBy('symbol').orderBy(F.col('timestamp'))
from pyspark.sql import functions as F

df_e = df_e.withColumn(
    "time_gap",
    (F.unix_timestamp(F.lead("timestamp").over(window)) 
     - F.unix_timestamp(F.col("timestamp"))) / 60
)
```

```python
count_timestamp = df_e\
                    .groupBy('symbol')\
                    .agg(
                        (F.sum(
                            F.when(
                                F.col('time_gap') == 1.0,1)
                            .otherwise(0))
                            /(F.count(F.col('time_gap')))).alias('timestamp_var'))
```

```python
count_timestamp.show()
```

```python
window = Window.partitionBy(F.col('symbol'),F.col('timestamp')).orderBy('timestamp')
dup_pct = df_e.withColumn('is_dup',F.row_number().over(window))
dup_pct = dup_pct.groupBy('Symbol')\
.agg(
    (1-(F.sum(
        F.when(
            F.col('is_dup')==1,1)
    .otherwise(0))/F.count(F.col('is_dup')))).alias('dup_pct'))
```

```python
dup_pct.show()
```

```python
df =df_e
window = Window.partitionBy(F.col('symbol')).orderBy('timestamp')
df = df.withColumn('return',F.lead(F.col('close'),1).over(window)-F.col('close'))\
       .withColumn('return_pct',F.col('return')/F.col('close'))\
       .withColumn('log_return',F.log(F.col('return_pct')))
                   
```

```python
df.show()
```
