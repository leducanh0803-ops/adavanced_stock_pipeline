```python
from pyspark.sql import SparkSession
 
spark = SparkSession.builder.appName("silver_check").getOrCreate()
 
# 1. List namespaces and tables
print("Namespaces:")
spark.sql("SHOW NAMESPACES IN nessie").show()
 
print("Tables in silver:")
spark.sql("SHOW TABLES IN nessie.silver").show()
 
# 2. Inspect the ohlcv table (skip if it doesn't exist)
table = "nessie.silver.ohlcv"
 
if spark.catalog.tableExists(table):
    print(f"\nSchema for {table}:")
    spark.sql(f"DESCRIBE TABLE {table}").show(truncate=False)
 
    print(f"\nRow count: {spark.table(table).count()}")
 
    print(f"\nSample rows:")
    spark.table(table).show(10, truncate=False)
 
    print(f"\nSnapshots / history:")
    spark.sql(f"SELECT * FROM {table}.history").show(truncate=False)
    spark.sql(f"SELECT * FROM {table}.snapshots").show(truncate=False)
else:
    print(f"{table} does not exist.")
 
spark.stop()
 
```
